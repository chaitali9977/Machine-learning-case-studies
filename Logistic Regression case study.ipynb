```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/insurance2.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
X=data.drop(['insuranceclaim'],axis=1) # here axis 1 means refering coloumn and axis 0 means rows
```

```python
X.head()
```

```python
y=data['insuranceclaim']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
y_pred=logmodel.predict(X_test)
```

```python
print(y_pred)
```

```python
#calculating accuracy of the Classification model
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
ac=accuracy_score(y_test,y_pred)
```

```python
print(cm) # 82=true positive,33=true negative,18=false positive,135=false negative
```
